```python
# K-Means Clustering using store_customers.csv

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Load dataset
data = pd.read_csv("store_customers.csv")

# Handle missing values by dropping rows with NaN
data.dropna(inplace=True)

# Step 2: Select features for clustering
# Using Annual Income and Spending Score
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Step 3: Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=0)

# Fit the model
kmeans.fit(X)

# Step 4: Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("\nCentroids:")
print(centroids)

# Step 5: Plot clusters
plt.scatter(X['Annual Income (k$)'],
            X['Spending Score (1-100)'],
            c=labels)

# Plot centroids
plt.scatter(centroids[:, 0],
            centroids[:, 1],
            marker='x',
            s=200)

plt.title("K-Means Clustering (Mall Customers)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.show()
```
